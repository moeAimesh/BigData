## Generator Functions

A generator function is defined like a regular function but uses the `yield` keyword to return data. Each call to `yield` produces the next value in the sequence, and the function's state is saved between each `yield`.

### Exercise 2: Write a prime number generator
+ Write a generator that will output one prime number at a time (using `yield`).

**Hint**: use a set (or list) to keep all prime numbers we found so far. This can be used to check if a new number is divisible by any of these.

+ Run this generator in a loop for about 10 seconds (> 10 seconds --> `break`). Which prime number did you reach in this amount of time?
+ Any idea how you can speed up this code?

```python
def prime_generator():
    """Generator that iterates through all prime numbers.
    The first prime number to be returned is 2.
    """
    prime_cache = [2, 3]
    yield 2
    yield 3
    n = 3
    while True:
        n += 2
        is_prime = True

        for prime in prime_cache:
            if n % prime == 0:
                is_prime = False
                break

        if is_prime:
            prime_cache.append(n)
            yield n

        
```

Once this test passes, let's have a look at how fast your implementation is.

Here we will do this by running the generator for a fixed amount of time.

```python
def prime_generator_speed_test(my_generator, time_to_stop=10):
    tstart = time.time()
    while True:
        if time.time() - tstart >= time_to_stop:
            break
    
        prime = next(my_generator)
    
    print(f"This generator ran for {(time.time() - tstart):.4f} seconds and reached the prime number: {prime}!")
```

## Generator Comprehension

Similar to the Python *list comprehension* there is also a *generator comprehension*, simply using round instead of squared brackets...

```python
squares = [x**2 for x in range(100)]
print(squares)
```

```python
import itertools

squares_gen = (x**2 for x in itertools.count(1))
print(squares_gen)
```
